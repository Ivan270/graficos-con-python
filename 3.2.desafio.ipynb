`````python
import matplotlib.pyplot as plt
import pandas as pd
`````

`````python
tiendas = ["A", "B", "C", "D"]
ventas_2022 = {
    "Ene": [100, 80, 150, 50],
    "Feb": [120, 90, 170, 60],
    "Mar": [150, 100, 200, 80],
    "Abr": [180, 110, 230, 90],
    "May": [220, 190, 350, 200],
    "Jun": [230, 150, 280, 120],
    "Jul": [250, 170, 300, 140],
    "Ago": [260, 180, 310, 150],
    "Sep": [240, 160, 290, 130],
    "Oct": [220, 140, 270, 110],
    "Nov": [400, 220, 350, 190],
    "Dec": [300, 350, 400, 250],
}
`````

`````python
df_ventas = pd.DataFrame(ventas_2022, index=tiendas)

fig, axs = plt.subplots(2, 2, figsize=(14, 8))

fig.subplots_adjust(hspace=0.6, wspace=0.4)

fig.suptitle("Variación de ventas en las 4 tiendas\nEne - Dic 2022", fontsize=14)
colors = ['darkviolet', 'green', 'cyan', 'darkblue']

ymin = 0
ymax = 450

for i, ax in enumerate(axs.flat):
    ax.plot(df_ventas.loc[df_ventas.index[i]], color=colors[i], lw=2)
    ax.set_title(f'Ventas: Tienda {df_ventas.index[i]}', loc='left')
    ax.set_xlabel('Mes', fontsize=12)
    ax.set_ylabel('Ventas', fontsize=12)
    ax.grid(linestyle=':')
    ax.tick_params(color='lightgrey')
    ax.set_ylim(ymin, ymax)
    ax.set_frame_on(False)

plt.show()
`````
