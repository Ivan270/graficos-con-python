# Conociendo los datos

```python
import pandas as pd

df = pd.read_csv('data/Inmigrantes_Canadá.csv')
df.head()
```

```python
df.shape
```

```python
# revisando datos y sus tipos
df.info()
```

```python
# Pais será el índice para facilitar lectura
df.set_index('Pais', inplace=True)
```

```python
df.head()
```

```python
# Obtener los inmigrantes de Chile

# Crear lista con años disponibles
anos = list(map(str, range(1980, 2014)))

chile = df.loc['Chile', anos]

# se generará un pandas series con el año como índice y la cantidad de migrantes en la primera columna
chile
```

```python
# Será mas conveniente crear un DataFrame con esta información.
# Se hará a través de un diccionario
cl_dict = {'Año': chile.index.tolist(), 
            'Inmigrantes': chile.values.tolist()}
```

```python
# Crear dataFrame a partir de diccionario
datos_cl = pd.DataFrame(cl_dict)
datos_cl.tail()
```

# Creando el primer gráfico con matplotlib

```python
import matplotlib.pyplot as plt

# Tamaño de la figura, en pulgadas (ancho, alto)
plt.figure(figsize=(8, 4))

plt.plot(datos_cl.Año, datos_cl.Inmigrantes)

# para mejorar la grfica en eje X y mostrar menos años
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])

# para que no aparezca mensaje de espacio de memoria donde se almacenó
plt.show()
```

# Añadiendo títulos y rótulos

```python
# Datos anteriores
plt.figure(figsize=(8, 4))
plt.plot(datos_cl.Año, datos_cl.Inmigrantes)
plt.xticks(["1980", "1985", "1990", "1995", "2000", "2005", "2010"])

# Para el título
plt.title('Inmigración de Chilenos hacia Canadá')

# Etiquetas eje x e y
plt.xlabel('Año')
plt.ylabel('Número dew inmigrantes')

plt.show() # debe ser el último comando!
```

En el gráfico se puede observar una cierta tendencia:
- Desde el año 1980 hasta 1985 disminuyó la inmigración desde un poco más de las 1200 hasta menos de 600
- Alcanzó un peak el año 1987 casi llegando a las 1600
- Hubo otro peak en 1991 casi llegando a los 1800
- Hubo una disminución fuerte de la inmigración hacia Canadá desde el entre 1990 y 1995 (aprox 1993-1994)
- Desde ahí se ha mantenido estable entre las 300 y 600 inmigraciones hasta el año 2013
